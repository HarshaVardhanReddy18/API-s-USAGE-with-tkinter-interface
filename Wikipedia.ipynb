```python
pip install wikipedia
import tkinter as tk
from tkinter import font
from tkinter import *
import wikipedia
import requests
from bs4 import BeautifulSoup
root = tk.Tk()
h = 800
w = 800

def display(entry):
    try:
        
        q = wikipedia.summary(str(entry),sentences = 5)
        
        label1['text'] = q
    except:
        url = 'https://google.com/' + str(entry)
        q = requests.get(url)
        q = q.text
        a = BeautifulSoup(q,'html.parser')
        label1['text'] = a
        
canvas = tk.Canvas(root,height = h, width = w,bg = '#33fff5')
canvas.pack()

frame = tk.Frame(root,bg = 'black')
frame.place(relx = .1, rely = .1,relheight = .8,relwidth = .8)

label = tk.Label(frame,bg = 'black',fg = 'red',text = 'WIKINANI',font = ('Gothic',30))
label.place(relx = .125 , rely = 0,relheight = .15,relwidth = .75)

label = tk.Label(frame,bg = 'black',fg = 'red',text = 'ENTER ABOUT ANYTHING THAT YOU WANT TO KNOW',font = ('Courier',15))
label.place(relx = .1 , rely = .1,relheight = .15,relwidth = .85)

entry = tk.Entry(frame,bg = 'white',fg = 'black',font = ('Courier',15))
entry.place(relx = .25,rely = .25,relheight = .05,relwidth = .5)

button = tk.Button(frame,text = 'GET INFO',bg = '#e60b1a',fg = 'yellow',font = ('Courier',15),command = lambda: display(entry.get()))
button.place(relx = .77,rely = .25,relheight = .05,relwidth = .2)

lowerframe = tk.Frame(frame,bg = 'white')
lowerframe.place(relx = .1,rely = .35,relheight =.6,relwidth = .8)

label1 = tk.Label(lowerframe,bg = 'white',fg = 'black' ,font = ('Gothic,10'),wraplength = 400)
label1.place(relwidth =1,relheight = 1)

scrollbar = tk.Scrollbar(lowerframe)
scrollbar.pack(side = 'right',fill = 'y')


lowerframe.yscrollcommand = scrollbar.set



scrollbar.config()



root.mainloop()
```
